### 재무 데이터 전처리 함수

```python
def get_finance_data(path):
    data_path=path
    raw_data=pd.read_excel(data_path)
    big_col=list(raw_data.columns)
    #big_col=big_col[1:len(big_col)]
    small_col=list(raw_data.iloc[0])
    small_col=small_col[1:len(small_col)]
    #big_col=list(raw_data.columns)
    raw_data=raw_data.set_index(big_col[0])
    big_col=big_col[1:len(big_col)]
    new_big_col=[]
    for num,col in enumerate(big_col):
        if 'Unnamed' in col:
            new_big_col.append(new_big_col[num-1])
        else:
            new_big_col.append(big_col[num])
    raw_data.columns=[new_big_col,small_col]
    clean_df=raw_data.loc[raw_data.index.dropna()]
    
    return clean_df
```

### PER 기준 오름차순 정렬하는 함수

```python
def low_per(invest_df,index_date,num):
    invest_df[(index_date,'PER')]=pd.to_numeric(invest_df[(index_date,'PER')])
    per_sorted=invest_df.sort_values(by=(index_date,'PER'))
    return per_sorted[index_date][:num]
```

```python
low_per(invest_df,'2019/12',10)
```

### ROA 기준 내림차순 하는 함수

```python
def high_roa(fr_df,index_date,num):
    fr_df[(index_date,'ROA')]=fr_df[(index_date,'ROA')].apply(check_IFRS)
    fr_df[(index_date,'ROA')]=pd.to_numeric(fr_df[(index_date,'ROA')])
    roa_sorted=fr_df.sort_values(by=(index_date,'ROA'),ascending=False)
    return roa_sorted[index_date][:num]
```

### 마법 공식 함수로 만들기

```python
def magic_formula(fr_df,invest_df,index_date,num):
    per=low_per(invest_df,index_date,None)
    roa=high_roa(fr_df,index_date,None)
    per['PER순위']=per['PER'].rank()
    roa['ROA순위']=roa['ROA'].rank(ascending=False)
    magic=pd.merge(per,roa,how='outer',left_index=True,right_index=True)
    magic['마법공식 순위']=(magic['PER순위']+magic['ROA순위']).rank().sort_values()
    magic=magic.sort_values(by='마법공식 순위')
    return magic[:num]
    
```

#### PER 말고도 PBR,PSR,PCR 같은 지표들 모두 '저평가된 주식'지표를 나타냄을 알게되었다
* PER: 순이익 대비 주가의 저평가 정도
* PBR: 순자산 대비 주가의 저평가 정도
* PCR: 자금 조달 능력 대비 주가의 저평가 정도
* PSR: 주당 매출액 대비 주가의 저평가 정도

#### 이를 고려하여 해당 지표들이 모두 낮은 주식을 생성하는 함수를 고려해볼 필요가 있어보였다

```python
def get_value_rank(invest_df,value_type,index_date,num):
    invest_df[(index_date,value_type)]=pd.to_numeric(invest_df[(index_date,value_type)])
    value_sorted=invest_df.sort_values(by=(index_date,value_type))[index_date] #기준 밸류,날짜 대상으로 정렬
    value_sorted[value_type+'순위']=value_sorted[value_type].rank() #랭크를 메겨 추가된 칼럼에 저장 
    return value_sorted[[value_type,value_type+'순위']][:num] #입력된 num 만큼의 순위 종목들 출력 
```

```python
get_value_rank(invest_df,'PSR','2019/12',20)
```

### 주식 저평가 지표 PER,PBR,PSR,PCR 모두 고려할 수 있는 함수 

```python
def mix_value(value_list,invest_df,index_date,num):
    for i,value in enumerate(value_list):  # 순서는 i, 적용 밸류는 value
        temp_df=get_value_rank(invest_df,value,index_date,None)
        if i==0: #첫 밸류 적용시(비어있는 데이터 프레임)
            mixed_value_df=temp_df
            rank=temp_df[value+'순위']
        else:
            mixed_value_df=pd.merge(mixed_value_df,temp_df,how='outer',left_index=True,right_index=True) #이전 밸류 적용 테이블에 합병
            rank=rank+temp_df[value+'순위']
    mixed_value_df['종합순위']=rank.rank() # 종합순위 칼럼 생성 
    mixed_value_df=mixed_value_df.sort_values(by='종합순위') #종합 순위 기준으로 데이터 테이블 정렬 
    return mixed_value_df[:num] #입력한 인자만큼 순위권 종목 출력 
```

### F_score
우량 주식을 찾는 대표적인 방법을, 다음의 조건을 설정한 후, 조건 충족시 1점을 주는 방식으로 총 점수를 계산하여 우량기업을 골라내는 방식이다  
* 당기 순이익 0 이상 여부
* 영업 현금흐름 0 이상 여부
* ROA 전년대비 증가 여부
* 영업현금흐름이 순이익보다 높은지 여부
* 부채비율 전년대비 감소 여부
* 유동비율 전년대비 증가 여부
* 당해 신규주식 발행여부(발행하지 않았을 경우 1)
* 매출 총이익 전년대비 증가 여부
* 자산 회전율 전년대비 증가 여부

#### 수집할 수 있는 데이터가 제한되어, 9가지의 평가 지표중 다음 3가지 지표만을 활용하였다(총점 3점)

* 당기 순이익 0 이상 여부
* 영업 현금흐름 0 이상 여부
* 영업현금흐름이 순이익보다 높은지 여부


### 3지표 F-score 함수:


```python
def get_fscore(fs_df,index_date,num):
    fscore_df=fs_df[index_date]
    fscore_df['당기순이익점수']=fscore_df['당기순이익']>0
    fscore_df['영업활동점수']=fscore_df['영업활동으로인한현금흐름']>0
    fscore_df['순이익보다 큰 영업활동점수']=fscore_df['영업활동으로인한현금흐름']>fscore_df['당기순이익']
    fscore_df['종합점수']=fscore_df[['당기순이익점수','영업활동점수','순이익보다 큰 영업활동점수']].sum(axis=1)
    fscore_df=fscore_df[fscore_df['종합점수']==3]
    return fscore_df[:num]
```
